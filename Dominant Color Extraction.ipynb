```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("elephant.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
orimg = img.shape
print(orimg)
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
pixels=img.reshape((330*500,3))
print(pixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=5)
km.fit(pixels)
```

```python
centers = km.cluster_centers_
centers = np.array(centers, dtype='uint8')
print(centers)
```

```python
plt.figure(0,figsize=(6,2))

i=1

colors = []
for each_col in centers:
    plt.subplot(1,5,i)
    plt.axis('off')
    i+=1
    
    colors.append(each_col)
    
   
    a = np.zeros((100,100,3), dtype='uint8')
    a[:,:,:] = each_col
    plt.imshow(a)

plt.show()
```

```python
nwimg = np.zeros((330*500, 3), dtype='uint8')

print(nwimg.shape)
```

```python
colors
```

```python
km.labels_
```

```python
for i in range(nwimg.shape[0]):
    nwimg[i] = colors[km.labels_[i]]
    

nwimg = nwimg.reshape((orimg))

plt.imshow(nwimg)
plt.axis("off")
plt.show()
```
